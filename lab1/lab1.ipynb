```python
import numpy as np
import scipy as sc
import math
import matplotlib.pyplot as plt
```

Фігури Лісажу 1-го роду

1. $\omega_{1} / \omega_{2} = 1$ і $\omega_{1} / \omega_{2} \approx 1$
2. $nx = cos(\omega_{1} t + \varphi), y = cos(\omega_{2} + \varphi)$

```python
w1 = [3, 3.1, 3.01, math.sqrt(2)]
w2 = [3, 2.9, 3.00, 1] 

delta = np.pi / 2
t = np.linspace( -np.pi * 4, np.pi * 4, 400)

for i in range(0,4):
 x = np.cos(w1[i] * t + delta)
 y = np.cos(w2[i] * t)
 plt.subplot( 2, 2, i + 1)
 plt.plot( x,y)

plt.show()
```

на 4 графіку відношення між частотами ірраціональне

повернемо фігури Лісажу. для цього змінимо $\Delta$ з $\pi / 2$ на $\pi / 2 + 0.5$

```python
delta = np.pi / 2 + 0.5

for i in range( 0, 4):
    x = np.cos(w1[i] * t + delta)
    y = np.cos(w2[i] * t)
    plt.subplot( 2, 2, i + 1)
    plt.plot( x,y )

plt.show()
```

Фігури Лісажу 2-го порядку

ситуація коли $\omega_{1} / \omega_{2} = 2$ і $\omega_{1} / \omega_{2} \approx 2$

```python
w1 = [6, 6, 6.01, 4.00]
w2 = [3, 3.1, 3.00, 1.99]

delta = np.pi / 2
t = np.linspace( -np.pi * 5, np.pi * 5, 500)

for i in range( 0, 4):
    x = np.cos( w1[i] * t + delta)
    y = np.cos( w2[i] * t)
    plt.subplot( 2, 2, i + 1)
    plt.plot( x,y )

plt.show()
```

Годограф Михайлова

поліном

$D(p) = p^{99} + p^{88} + p^{77} + p^{5} + p + 4$

```python
from sympy import *
w = symbols('w', real=True)

D = pow(I*w, 99) + pow(I*w, 88) + pow(I*w, 77) + pow(I*w, 5) + I*w + 4
print(D)

D_Re = re(D)
D_Im = im(D)


x=[D_Re.subs({w:n}) for n in np.arange(0,100,0.1)]
y=[D_Im.subs({w:n}) for n in np.arange(0,100,0.1)]

plt.axis([-300, 300, -1000, 1000])
plt.plot(x, y)
plt.grid(True)
plt.show()
```
